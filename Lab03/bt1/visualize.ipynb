## Trực quan hóa dữ liệu

### 1. Histogram

```python
cols = df.columns.tolist()
n_cols = 2
n_rows = (len(cols) + 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, 4 * n_rows))
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.histplot(df[col], ax=axes[i], kde=True)
    axes[i].set_title(f'Distribution of "{col}"')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')

# Ẩn các subplot thừa nếu số cột lẻ
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```

**Kết quả phân tích**
1. Fixed acidity

* Phân phối lệch phải nhẹ, tập trung ở khoảng **6–8**.
* Ít giá trị ngoại lai trên 12.

2. Volatile acidity

* Gần đối xứng, đỉnh ở **0.4–0.6**.
* Một số giá trị ngoại lai trên 1.0.

3. Citric acid

* Lệch phải, nhiều mẫu ở **0.0–0.2**.
* Ít trường hợp có giá trị cao >0.6.

4. Residual sugar

* Lệch phải mạnh, chủ yếu ở **2–3**.
* Ngoại lai trên 12 nhưng hiếm.

5. Chlorides

* Lệch phải, tập trung quanh **0.05–0.1**.
* Một số ngoại lai >0.3.

6. Free sulfur dioxide

* Lệch phải, tập trung nhiều ở **5–20**.
* Có giá trị cao tới 70.

7. Total sulfur dioxide

* Lệch phải mạnh, nhiều mẫu <100.
* Ngoại lai gần 300.

8. Density

* Gần phân phối chuẩn, đỉnh **0.996–0.997**.
* Trải rộng từ 0.990–1.004.

9. pH

* Gần chuẩn, tập trung ở **3.2–3.4**.
* Ngoại lai ít (<3.0 hoặc >3.6).

10. Sulphates

* Lệch phải, nhiều mẫu **0.5–0.7**.
* Một số ngoại lai >1.5.

11. Alcohol

* Lệch phải, tập trung **9–11**.
* Có mẫu cao tới 14–15.

12. Quality

* Biến rời rạc, phân phối mất cân bằng.
* Tập trung ở **5–6**, ít ở mức 3, 4, 7, 8.

---

**Nhận xét**

* Nhiều biến có phân phối **lệch phải** (residual sugar, chlorides, sulfur dioxide, alcohol).
* Một số biến gần chuẩn: **density, pH, fixed acidity**.
* **Quality** mất cân bằng, chủ yếu ở mức 5–6.

### 2. ScatterPlot

```python
target = 'quality'
features = [col for col in df.columns if col != target]
n_cols = 2
n_rows = (len(features) + 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, 4 * n_rows))
axes = axes.flatten()

for i, col in enumerate(features):
    sns.scatterplot(x=df[col], y=df[target], ax=axes[i], hue=df[target])
    axes[i].set_title(f'Relationshop between "{col}" and "{target}"')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel(target)

for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```

**Mối quan hệ giữa các biến và chất lượng rượu**

 1. Biến có ảnh hưởng mạnh

* **Alcohol**: Rượu chất lượng cao (7–8) thường có **độ cồn cao hơn (11–13)**. Rượu kém chất lượng tập trung ở mức thấp (8–10).
* **Volatile acidity**: Chất lượng cao đi kèm với **acid bay hơi thấp**. Rượu chất lượng kém thường có volatile acidity >0.6.
* **Sulphates**: Rượu tốt thường có **sulphates cao hơn (0.6–1.0+)**.
* **Citric acid**: Rượu ngon có xu hướng chứa nhiều citric acid, trong khi rượu kém chất lượng thường gần 0.

 2. Biến có ảnh hưởng vừa

* **Chlorides**: Hàm lượng chloride cao thường xuất hiện ở rượu chất lượng thấp.
* **Density**: Rượu có mật độ cao thường kém chất lượng; mật độ thấp hơn liên quan đến rượu tốt.

 3. Biến ảnh hưởng yếu hoặc không rõ ràng

* **Fixed acidity, residual sugar, free sulfur dioxide, total sulfur dioxide, pH**: Không thể hiện xu hướng rõ ràng với chất lượng.

---

**Kết luận**

* **Yếu tố tích cực:** Alcohol, sulphates, citric acid.
* **Yếu tố tiêu cực:** Volatile acidity, chlorides, density.
* Các biến khác đóng vai trò phụ hoặc ít ảnh hưởng.

Điều này cho thấy rượu vang ngon thường có:

* Độ cồn cao, sulphates và citric acid hợp lý,
* Ít acid bay hơi, ít chloride, mật độ thấp.


### 3. Heatmap

```python
plt.figure(figsize=(12, 8))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

**Phân tích Correlation Heatmap**

Heatmap trên biểu diễn **ma trận tương quan (correlation matrix)** giữa các biến hóa học trong rượu vang và chất lượng (`quality`).

 1. Các biến có tương quan mạnh với **quality**

* **Alcohol (0.48, dương)**

  * Chất lượng rượu tăng khi độ cồn cao → yếu tố quan trọng nhất.
* **Volatile acidity (-0.39, âm)**

  * Acid bay hơi cao làm giảm chất lượng rượu.
* **Sulphates (0.25, dương)**

  * Hàm lượng sulphates cao hơn thường liên quan đến rượu chất lượng cao.
* **Citric acid (0.23, dương)**

  * Tăng citric acid giúp cải thiện chất lượng rượu.

 2. Các biến có tương quan yếu hoặc gần như không ảnh hưởng

* **Fixed acidity (0.12)**: Gần như không liên quan nhiều đến chất lượng.
* **Residual sugar (0.01)**: Không ảnh hưởng rõ ràng.
* **Chlorides (-0.13)**: Tương quan rất yếu, có thể ảnh hưởng nhẹ tiêu cực.
* **Free sulfur dioxide (-0.05)**, **Total sulfur dioxide (-0.19)**: Ảnh hưởng yếu.
* **Density (-0.17)**: Có xu hướng âm, rượu đặc hơn thường kém chất lượng.
* **pH (-0.06)**: Không đáng kể.

 3. Các mối quan hệ mạnh giữa các biến độc lập

* **Fixed acidity & citric acid (0.67)**: Hai biến này thường tăng/giảm cùng nhau.
* **Free sulfur dioxide & total sulfur dioxide (0.67)**: Quan hệ hiển nhiên (tổng và thành phần).
* **Density & fixed acidity (0.67)**: Rượu có độ acid cao thì mật độ cũng thường cao.
* **Density & alcohol (-0.50)**: Rượu có nhiều cồn thường ít đặc (mật độ thấp hơn).
* **pH & fixed acidity (-0.68)**: Quan hệ ngược chiều, pH thấp khi acid cao.

---

**Kết luận**

* **Yếu tố cải thiện chất lượng:** Alcohol, sulphates, citric acid.
* **Yếu tố làm giảm chất lượng:** Volatile acidity, chlorides, density.
* Các biến khác ảnh hưởng yếu, ít quan trọng trong dự đoán chất lượng.
